## Singlet triplet difference ionization Energies

```python
# Read data
df = pd.read_csv('sample_data/BPE.csv', sep = '\s+')
# Het it into arrays
x2 = np.array(df['x2'])
y2 = np.array(df['y2'])
```

```python
# Read data
df_2 = pd.read_csv('sample_data/BPDE.csv', sep = '\s+')
# Het it into arrays
x = np.array(df_2['x'])
y = np.array(df_2['y'])
```

```python
# Plotting Difference Ionization Energies
with plt.style.context(['science', 'notebook']):
    
    plt.figure(figsize = (10,5))
    
    plt.plot(x2, y2, linestyle = '-.', lw = 2, marker = 'o', c = 'b',  alpha = 0.5, label = 'BPE Method')
    plt.scatter(x, y, marker = 'o', lw = 2, c = 'm', alpha = 0.5, label = 'BPDE Method')

    
    plt.xlabel(r'$\rm R(H\cdots H)/ A^$')
    plt.ylabel(r'$\rm \Delta E_{ST}(S)\, - \,\Delta E_{ST}(Cl)\quad [kcal\, mol^{-1}]$')
    plt.legend()
    plt.grid(ls = ':')
    
    plt.savefig("./Figures/ionization_energies.svg".format())

    plt.show()
```
